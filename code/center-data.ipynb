# Centrar datos

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd                # trabajar bases de datos (DataFrame)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

## Read data

```python
df = sns.load_dataset('penguins')       # importar el conjunto de datos
```

```python
df.shape
```

## DataFrame exploration

```python
df
```

```python
df = df.dropna()
```

```python
data = df[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']]
data
```

```python
X = data.values
X
```

```python
data.mean()
```

```python
mu = X.mean(axis=0)
mu
```

```python
sig = X.std(axis=0)
```

```python
# centered data
Xc = X - mu

# normalized data
Xn = (X - mu) / sig
```

```python
data_centered   = pd.DataFrame(data=Xc, columns=data.columns)
data_normalized = pd.DataFrame(data=Xn, columns=data.columns)
```

```python
data_centered
```

```python
data_normalized
```

```python
plt.figure(figsize=(20,6))
sns.pairplot(data=data, diag_kind='kde')
plt.show()
```

```python
plt.figure(figsize=(20,6))
sns.pairplot(data=data_centered, diag_kind='kde')
plt.show()
```

```python
plt.figure(figsize=(20,6))
sns.pairplot(data=data_normalized, diag_kind='kde')
plt.show()
```
